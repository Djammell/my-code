```python
import gym
import numpy as np, pandas as pd
```

```python
env = gym.make('Taxi-v2')
env.reset()
env.render()
```

```python
nStates = env.observation_space.n
nActions = env.action_space.n
```

```python
policy = np.ones((nStates, nActions))/nActions
```

```python
def generateSession(iterationsPerSession = 10**4, gamma = 0.3, percentile = 0, nSamples = 0):
    states, actions = [], []
    totalReward = 0
    
    stateS0 = env.reset()
    
    for i in range(iterationsPerSession):
        action = np.random.choice(nActions, p = policy[stateS0,:,percentile, nSamples])
        
        stateS1, reward, done, _ = env.step(action)
        
        states.append(stateS0)
        actions.append(action)
        totalReward += reward
        
        stateS0 = stateS1
        
        if done == True:
            break
    return states, actions, totalReward
```

# HOMEWORK 1.1
playing with different values of percentile and samples count

```python
nSamples = 150
percentile = 30
smoothing = 0.2
nSessions = 150

allRewardsBySession = [] #награды за каждую сессию

for numS, nSamples in enumerate(range(250,350,50)):
    allRewardsByPercentile = [] #значение награды в кажой вариации персентиля 
    policy = np.ones((nStates, nActions,len(range(30,60,10)),len(range(250,450,50))))/nActions
    
    for num,percentile in enumerate(range(30,60,10)):
        rewardsHistory = [] # 
        print('samples count = {} and percentile {}'.format(nSamples,percentile))
        for i in range(nSessions):
            sessions = (generateSession(percentile = num, nSamples= numS) for i in range(nSamples))
        
            batchStates, batchActions, batchRewards = map(np.array, zip(*sessions))
    
    
            threshold = np.percentile(batchRewards, percentile)
    
            eliteStates = batchStates[batchRewards > threshold]
            eliteActions = batchActions[batchRewards > threshold]
    
            eliteStates, eliteActions = map(np.concatenate, [eliteStates, eliteActions])
    
            elitePolicy = np.zeros_like(policy[:,:,num, numS]) + smoothing
    
            for j in range(len(eliteStates)):
                elitePolicy[eliteStates[j], eliteActions[j]] += 1
        
            #нормализуем политику чтобы в колонках не было значений больше 1
            policy[:,:,num, numS] = elitePolicy/elitePolicy.sum(1, keepdims = True)
            rewardsHistory.append(np.mean(batchRewards))# значение средней награды за сессию их будет 150
#             print('mean reward - {}, threshold - {}'.format(np.mean(batchRewards),threshold))        
        
        print('mean reward - {}'.format(np.mean(batchRewards))) 
        allRewardsByPercentile.append(rewardsHistory)
    allRewardsBySession.append(allRewardsByPercentile)
    
```
